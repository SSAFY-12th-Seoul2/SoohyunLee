```python
def min_distance(dg_d, dg_p, store_d, store_p):
    if dg_d in [1, 2]:
        if store_d in [1, 2]:
            # 동근이랑 상점의 위치가 같을 경우
            if dg_d == store_d:
                distance = abs(dg_p - store_p)
            else:
                # 동근이랑 상점의 위치가 반대편인 경우
                distance = min(dg_p + height + store_p, (width-dg_p) + height + (width - store_p))
        else:
            # 동근이가 북쪽에 있을때
            if dg_d == 1:
                # 상점의 위치가 서쪽인 경우
                if store_d == 3:
                    distance = dg_p + store_p
                # 상점의 위치가 동쪽인 경우
                else:
                    distance = (width - dg_p) + store_p
            else:
                ## 동근이가 남쪽에 있을 때
                # 상점이 서쪽에 있을 때
                if store_d == 3:
                    distance = dg_p + (height - store_p)
                else:
                    # 상점이 동쪽에 있을 때
                    distance = (width - dg_p) + (height - store_p)

    else:
        if store_d in [3, 4]:
            # 동근이랑 상점의 위치가 같은 경우
            if dg_d == store_d:
                distance = abs(dg_p - store_p)
            else:
                # 동근이랑 상점의 위치가 반대인 경우
                distance = min(dg_p + width + store_p, (height - dg_p) + width + (height - store_p))
        else:
            # 동근이가 서쪽에 있을 때
            if dg_d == 3:
                # 상점이 북쪽에 있는 경우
                if store_d == 1:
                    distance = dg_p + store_p
                # 상점이 남쪽에 있는 경우
                else:
                    distance = (height - dg_p) + store_p
            else:
                # 상점이 북쪽에 있는 경우
                if store_d == 1:
                    distance = dg_p + (width - store_p)
                # 상점이 남쪽에 있는 경우
                else:
                    distance = (height - dg_p) + (width - store_p)
    return distance

width, height = map(int, input().split()) 
store_num = int(input())
stores = [list(map(int,input().split())) for _ in range(store_num)]
dg_d, dg_p = map(int, input().split())

distance = 0
for store_d, store_p in stores:
    distance += min_distance(dg_d, dg_p, store_d, store_p)
print(distance)
```

```python
def min_distance(dg_d, dg_p, store_d, store_p, width, height):
    if dg_d == store_d:
        return abs(dg_p - store_p)

    elif dg_d in [1, 2] and store_d in [1, 2]:
        return min(dg_p + height + store_p, (width - dg_p) + height + (width - store_p))

    elif dg_d in [3, 4] and store_d in [3, 4]:
        return min(dg_p + width + store_p, (height - dg_p) + width + (height - store_p))

    elif dg_d == 1:
        return store_p + (dg_p if store_d == 3 else width - dg_p)

    elif dg_d == 2:
        return (height - store_p) + (dg_p if store_d == 3 else width - dg_p)

    elif dg_d == 3:
        return store_p + (dg_p if store_d == 1 else height - dg_p)

    elif dg_d == 4:
        return (width - store_p) + (dg_p if store_d == 1 else height - dg_p)

width, height = map(int, input().split()) 
store_num = int(input())
stores = [list(map(int, input().split())) for _ in range(store_num)]
dg_d, dg_p = map(int, input().split())

distance = sum(min_distance(dg_d, dg_p, store_d, store_p, width, height) for store_d, store_p in stores)
print(distance)
```
